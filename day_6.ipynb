```python
# Example:

def my_function():
  print("Hello from a function")

my_function()
```

```python
def is_float(x):
    if (type(x) == float or type(x) == int):
        print("It's a float.")
    else:
        print(f"It's not a float. It's a {type(x)}")
    return

number1 = 3.0
number2 = 6
word1 = "name"

is_float(number1)
is_float(number2)
is_float(word1)
```

```python
# Default Parameter Value:

def my_function(country = "Norway"):
  print("I am from " + country)

my_function("Sweden")
my_function("India")
my_function()
my_function("Brazil")
```

```python
# To let a function return a value, use the return statement:

def my_function(x):
  return 5 * x

print(my_function(3))
print(my_function(5))
print(my_function(9))
```

Iterating efficiently over a list <br>
Source:

Syntax:

list = [expression for local_variable in object]

This is equivalent to:

list = []
for local_variable in object:
    list.append(edth="400"/>


```python
# List comprehensions provide a concise way to create lists.

squares = [n**2 for n in range(10)]
squares
```

```python
# A list comprehension consists of brackets containing an expression followed by a for clause,
# then zero or more for or if clauses.

planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']

short_planets = [planet for planet in planets if len(planet) < 6]
short_planets
```

```python
# Example of list comprehensions:

import random as rd

n = 20
rand_num = [rd.randint(1,100) for _ in range(n)]
print(rand_num)
```

```python
# Filtering list comprehensions ~ using if condition:

filtered_squares = [n**2 for n in range(10) if (n**2)%2 != 0]
filtered_squares
```

```python
# Filtering list comprehensions ~ using if-else condition:

filtered_squares = [n**2  if (n**2)%2 != 0 else -1 for n in range(10)]
filtered_squares
```

```python
# Using zip ~ iterating over 'n' collections:

names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]
result = list(zip(names, ages))
print(result)
type(result[0])

colleagues = [(i,j) for (i,j) in zip(names,ages)]
```

```python
# Nested List Comprehensions:

A = ["a", "b",]
B = [10, 20, 30, 40]

prod_cart = {(a,b) for a in A for b in B}
print(prod_cart)
type(prod_cart)
```

```python
# Nested List Comprehensions:

A = ["a", "b",]
B = [10, 20, 30, 40]

prod_cart2 = {a:b for a in A for b in B}
print(prod_cart2)
type(prod_cart2)
```
