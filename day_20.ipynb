In this and the next class, we will begin our introduction to **Probability and Statistics**.

These two areas are extremely broad and provide the entire foundation of **data science**. Therefore, let's take a look at the main concepts. Probability and statistics can be divided into the following three pillars:

1 - **Probability**: It is the mathematical theory used to study the uncertainty arising from `random phenomena`.

<div style="background-color: lightgray; padding: 10px;">
    <h2> Median
</div>

The **median** of a variable is a number such that there are an equal number of observations greater and smaller than it, thus occupying the **central position** in the series of observations.

Therefore, to calculate the median, we simply **order the data from smallest to largest** and count what is in the middle of the ordered data (if there is an even number of data, take the average between the data that are "in the middle").

- *Example*

i) 3, 4, **7**, 8, 8 **median = 7**       (ODD CASE)

ii) 3, 4, **7**, **8**, 8, 9 **median = (7+8)/2 = 7.5**   (EVEN CASE)

Therefore, we can define:

- Median of $X = X_{\frac{n+1}{2}}$ `If sample size is odd`


- Median of $X = \frac{X_{\frac{n}{2}}+ X_{\frac{n+1}{2}}}{2}$ `If sample size is even`

Since **Python index values** start at `0`, we must note that the equation above should be as follows:

- Median of $X = X_{\frac{n}{2}}$ `If sample size is odd`


- Median of $X = \frac{X_{\frac{n-1}{2}}+ X_{\frac{n}{2}}}{2}$ `If sample size is even`

````python
# Demonstration:

values = [15, 3, 7, 1, 4] # len = 6 / 2 = 3
sorted_values = sorted(values)

median_index = int(len(sorted_values) / 2)
median = sorted_values[median_index]

if len(sorted_values) % 2 == 0:
  median += sorted_values[median_index - 1]
  median /= 2

print('Values:', sorted_values)
print('Median:', median)
````

#### Quartiles (Q1, Q2, and Q3)

These are values obtained from the dataset ordered in ascending order, dividing the data **into four equal parts**.

- The first quartile, Q1, is the number that leaves 25% of the observations below and 75% above.
- The second quartile, Q2, is the **median**, leaving 50% of the observations below and 50% above.
- The third quartile, Q3, leaves 75% of the observations below and 25% above.

#### Box Plot

The box plot is a graph used to assess the empirical distribution of data, giving us a **sense of how the data is distributed**. The box plot consists of the first and third quartiles and the median. The lines known as lower and upper whiskers extend, respectively, from the lower quartile to the smallest value not less than the lower limit and from the upper quartile to the largest value not greater than the upper limit. The limits are calculated as follows:

Lower limit: $ \max\{\min(\text{data});Q_1-1.5(Q_3-Q_1)\} $.

Upper limit: $ \min\{\max(\text{data});Q_3+1.5(Q_3-Q_1)\} $.

In this case, points outside these limits are considered outliers and are denoted by an asterisk (*). The following figure presents an example of the format of a box plot.

<img src=https://miro.medium.com/max/8000/1*0MPDTLn8KoLApoFvI0P2vQ.png width=500>

The box plot can also be used for visual comparison between two or more groups. For example, two or more boxes are placed side by side, and the variability between them, the median, and so on, are compared. Another important point is the difference between the quartiles $ (Q_3 - Q_1) $, which is a measure of data variability.

*But how to discover if my samples have a normal distribution?*

Discovering if your sample follows a normal distribution is important for many statistical analyses because several statistical tests and methods assume that the data are normally distributed. Here's a simplified explanation:

**Statistical Tests** to Explore Further:

- **Shapiro-Wilk Test**: This test calculates a W statistic that measures the discrepancy between the observed data and what would be expected under the assumption of normality. If the p-value from this test is greater than a chosen significance level (commonly 0.05), you fail to reject the null hypothesis that the data are normally distributed.
- **Kolmogorov-Smirnov Test**: This test compares the cumulative distribution function of your data with that of a normal distribution. If the p-value is greater than your chosen significance level, you fail to reject the null hypothesis that the data come from a normal distribution.
- **Anderson-Darling Test**: This is a modification of the Kolmogorov-Smirnov test that gives more weight to the tails of the distribution. Similar to the other tests, if the p-value is greater than your chosen significance level, you fail to reject the null hypothesis of normality.

These tests provide statistical evidence about whether your sample data likely come from a normal distribution. However, it's important to note that no test can definitively prove that your data are normally distributed. They can only provide evidence for or against the assumption of normality based on the sample data you have. Additionally, the appropriateness of these tests can depend on factors such as sample size and the presence of outliers.

<div style="background-color: lightgray; padding: 10px;">
    <h2> Variance
</div>


Variance represents **how much the data deviates from the mean, generally**. By definition, we take the average of the **squared deviations from the mean**:

$$ Var(X) = \sum_{i=1}^{n}\frac{(x-\overline{x})^{2}}{n-1} = \frac{(x_1-\overline{x})^{2}}{n-1} + \frac{(x_2-\overline{x})^{2}}{n-1} + \cdots$$, population variance (n) and sample variance (n - 1)

````python
print('Varincia amostral:', df_tips['size'].var())
print('Varincia populacional:', df_tips['size'].var(ddof=0))
````

<div style="background-color: lightgray; padding: 10px;">
    <h2> Standard Deviation
</div>


Note that, as defined, variance is a **quadratic** value in the data. To avoid this, it is common to take the square root of the variance, which defines the **standard deviation**:

$$ \sigma(X) \equiv \sqrt{Var(X)} = \sqrt{\sum_{i=1}^{n}\frac{(x-\overline{x})^{2}}{n - 1}}$$

````python
print('Desvio Padro amostral:', df_tips['size'].std())
print('Desvio Padro populacional:', df_tips['size'].std(ddof=0))
````

#### Pearson Correlation

[__Pearson Correlation__](https://en.wikipedia.org/wiki/Pearson_correlation_coefficient) is expressed as the **rate of linear relationship between two numeric variables**. The closer it is to 1 or -1, the more **linearly correlated the tested variables are**. The formula is defined as:

$r = \displaystyle{\frac{s_{XY}}{\sqrt{s^2_X s^2_Y }}}$

Property:

 $-1 \leq r \leq 1$

Common classifications are:

1. $r=1$ indicates a perfect positive correlation
2. $r=-1$ indicates a perfect negative correlation
3. $0.7 \leq |r| \leq 1$ indicates a strong correlation
4. $0.5 \leq |r| \leq 0.69$ indicates a moderate correlation
5. $0 \leq |r| \leq 0.49$ indicates a weak corre#lation
<br>

### Spearman Correlation

[__Spearman Correlation__](https://en.wikipedia.org/wiki/Spearman%27s_rank_correlation_coefficient) is another well-known linear correlation, and this coefficient indicates whether the samples have a *monotonic* correlation, meaning **if one increases or decreases, the other will increase or decrease**.

The coefficient varies between -1 and 1:

> The closer the coefficient is to 1, the more positively correlated the samples are, meaning if one increases, the other will increase (or if one decreases, the other will decrease).

> The closer the result is to -1, the more negatively correlated the variables are, meaning if one increases, the other will decrease (or if one decreases, the other will increase).

> A correlation of 0 indicates no correlation.

**Important:** the Spearman coefficient **does not** necessarily indicate linear correlation!

<div style="background-color: lightgreen; padding: 10px;">
    <h2> Skewness
</div>


It is the **degree of deviation or deviation** from the **symmetry** of a distribution. When the curve is symmetrical, the `mean`, the `median`, and the `mode` coincide at the same point, indicating a perfect balance in the distribution. An example of a symmetrical distribution is the **normal distribution**. When balance does not occur, the mean, median, and mode fall on different points of the distribution, and the distribution becomes **asymmetrical**:

<img src="https://www.researchgate.net/profile/Marco_Faria/publication/294890337/figure/fig5/AS:668873900716049@1536483373457/a-Negative-skewness-b-Normal-curve-c-Positive-skewness-Durkhure-and-Lodwal-2014.ppm" width=600>

- **Negative or left-skewed** asymmetrical distribution: when values concentrate on the upper end of the scale and gradually distribute towards the lower end.
  
- **Positive or right-skewed** asymmetrical distribution: when values concentrate on the lower end of the scale and gradually distribute towards the upper end.

The measure of asymmetry is known as **skewness**, calculated as:

$$ s(X) = \frac{1}{\sigma^3}\sum_{i=1}^{n}\frac{(x-\overline{x})^{3}}{n}$$

The value of skewness indicates the asymmetry of the distribution:

- s = 0: symmetrical distribution
- s > 0: right skewness
- s < 0: left skewness

### Frequency Tables

Frequency distribution is a grouping of data into classes, such that we count the number of occurrences in each class. The goal is to present the data in a more concise manner that allows us to extract information about its behavior. Below, we present some definitions necessary for constructing the frequency distribution.

**Absolute frequency ($_i$):** It is the number of observations corresponding to each class. The absolute frequency is generally referred to simply as frequency.

**Relative frequency ($_{ri}$):** It is the quotient between the absolute frequency of the corresponding class and the sum of frequencies (total observed), i.e., $_{ri}=\frac{_i}{\sum_{j}_j}$ where $n$ represents the total number of observations.

**Percentage frequency ($p_i$):** It is obtained by multiplying the relative frequency by 100%.
