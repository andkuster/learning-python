<div style="background-color: lightgreen; padding: 10px;">
    <h2> Line and Scatter Plot
</div>

The very first thing that we are going to display is the so-called `line plot` and `scatter plot`:

- the first one works by connecting two dots with a line;
- the second one simply draws all the dots we provide without any connecting lines;

Now letâ€™s create some dummy data to demonstrate how ``plt.plot()`` and ``plt.scatter()`` work.

```python
import numpy as np
import random

# Sample data for sales
days = np.arange(1, 11)
print(days)

sales_a = [random.randint(1, 10) + 10 for _ in range(10)]  # Generating random prices for product A
print(sales_a)
```

As we have already got our ``x-y`` pairs, we can simply run the following code to draw a line plot and scatter plot:

```python
plt.plot(days, sales_a)
```

```python
plt.scatter(days, sales_a)
```

---
We can edit some features in lineplots, like the `linestyle` and `linewidth`.

Some arguments possible to be used for this parameter is listed as follows:
- dotted (:)
- dashed (--)
- dashdot (-.)

```python
plt.plot(days, sales_a, linestyle='--', linewidth=0.5)
```

---
Now, with ``plt.scatter()``, you can change the dot style by specifying the **marker** parameter. If you omit this parameter as we did before, the default marker shape is a circle (â€˜oâ€™). Alternatively, you can use one of the following characters:
- `.` (dot)
- `p` (pentagon)
- `+` (thin plus symbol)
- `P` (thick plus symbol)
- `^` (triangle)
- `v` (upside down triangle)
- `<` (triangle pointing to the left)
- `>` (triangle pointing to the right)"
- `*` (star)
- `x` (thin cross sign)
- `X` (thick cross sign)
- `s` (square)

```python
plt.scatter(days, sales_a, marker='s', s=40)
```

Combining both techniques:

```python
plt.plot(days, sales_a, linestyle='dashdot')
plt.scatter(days, sales_a, marker='s', s=40)
```

**Tip:** to get rid of the output codes, we have to use ``plt.show()``.

```python
plt.plot(days, sales_a, linestyle='--')
plt.scatter(days, sales_a,)
plt.show()
```

    <h2> The f(x) Function
</div>

Now, we are going to learn how to draw a line based on a mathematical function. <br>
However, I want to introduce you a Numpy function of ``np.linspace()`` beforehand as it is going to be extremely necessary for the subsequent cases. 

This function essentially works by accepting **3 main input params**: start, stop and num. <br>
After being run, it will return a Numpy array of length num within the range of start and stop, in which the intervals between each two elements are uniform. 

```python
import numpy as np

np.linspace(0, 10, 5)
```

```python
np.linspace(-10, 10, 21)
```

Using this concept, we can create a curve using quadratic equation. In the following code block, weâ€™ll generate a curve following the equation y = xÂ².

```python
x = np.linspace(-10, 10, 42)
y = x**2

plt.plot(x, y)
plt.show()
```

Another example using a polinomial equation:

```python
x = np.linspace(-10, 10, 50)
y = x**3 + 2*x**2 - x + 8

plt.plot(x, y)
plt.show()
```

---
If you want to take a closer look, you need to use ``plt.xlim()`` and ``plt.ylim()`` to limit the range of the figure. 

Let's try to set the horizontal axis range to -5 and 5, while the vertical axis to -20 to 20 only. 

The output shown below represents the same curve as the one we saw above except that it only covers smaller region:

```python
x = np.linspace(-10, 10, 50)
y = x**3 + 2*x**2 - x + 8

plt.plot(x, y)
plt.xlim(-5, 5)
plt.ylim(-20, 20)
plt.show()
```

<div style="background-color: lightgray; padding: 10px;">
    <h2> Trigonometric functions
</div>

In this section let's see how to display the curve of a sine function. 

The way to do so is similar to the previous ones, except that here the y values are computed automatically using ``np.sin()``

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)

plt.plot(x, y1)
plt.show()
```

We are actually able to display multiple plots in a single figure. 

Let's draw a cosine wave alongside the sine one. 

The way to do it is easy, just put both ``plt.plot()`` together in the same notebook cell as what we did when stacking ``plt.plot()`` and ``plt.scatter()``.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

---
We can do actually a lot more we can do when it comes to creating Matplotlib graphs. 

Below, let's add figure title, axis labels, gridlines, and plot legends to the graph. We can also change the plot color by passing an argument for the c parameter (which it stands for â€œcolorâ€) and adjust the alpha transparency parameter:

```python
plt.plot(x, y1, c='green', alpha = 0.4)      
plt.plot(x, y2, c='#990F22', alpha = 1)    

#plt.title('Sine and Cosine Waves')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
#plt.grid()
plt.legend(['sin', 'cos']) #the list have to be defined in the correct order
plt.show()
```

<div style="background-color: lightgreen; padding: 10px;">
    <h2> Subplots
</div>

Instead of calling ``plt.show()`` repeatedly to display multiple figures, Matplotlib provides us with the so-called **subplot** to do so. 

The use of **subplot** itself might be somewhat more difficult at first, yet it allows us to have better flexibility. We are going to initialize some x-y pairs before actually drawing some graphs:

```python
x = np.linspace(0, 5, 100)

y1 = np.sin(np.cos(x**2))
y2 = np.cos(np.sin(x**2))
y3 = np.log(np.sin(x))
y4 = np.cos(np.sqrt(x))
y5 = x**2
y6 = np.sin(x**3)
```

As the above codeblock has been run, now that we have 6 arrays containing y values coming from different mathematical function outputs. 

Below we are going to plot them all in different subplots:

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 11))    #(1)
plt.suptitle('Drawing Multiple Figures')

axes[0,0].set_title('[0,0]: sin(cos(x**2))')    #(2)
axes[0,0].plot(x, y1, c='red')

axes[0,1].set_title('[0,1]: cos(sin(x**2))')
axes[0,1].plot(x, y2, c='green')

axes[1,0].set_title('[1,0]: log(sin(x))')
axes[1,0].plot(x, y3, c='pink')
axes[1,0].set_ylabel('output')    #(3)

axes[1,1].set_title('[1,1]: cos(sqrt(x))')
axes[1,1].plot(x, y4, c='blue')

axes[2,0].set_title('[2,0]: x**2')
axes[2,0].plot(x, y5, c='orange')
axes[2,0].grid()

axes[2,1].set_title('[2,1]: sin(x**3)')
axes[2,1].plot(x, y6, c='black')
axes[2,1].set_xticks([])
axes[2,1].set_yticks([])      #(4)

plt.show()
```

The line labeled as #(1) specifies that we want to create 6 figures arranged in 3 rows and 2 columns. 

Each subplot we created can be accessed using an array-like indexing method. For example:
- axes[0,0] refers to the top left subplot (red line);
- axes[0,1] refers to the subplot located at the top right position (green line);
- axes[1,0] is for the subplot at row 1 column 0 (pink line), and so forth;

Each of these subplots can actually be treated similarly to the standard figures we discussed previously. 

However, there are several differences that we need to pay attention to:
- Rather than using ``plt.title()`` for adding figure title, we need to use ``set_title()`` instead (#2);
- The same thing also applies to ``plt.ylabel()`` and ``plt.yticks()`` which need to be replaced by ``set_ylabel()`` (#3) and ``set_yticks()`` (#4) respectively;
